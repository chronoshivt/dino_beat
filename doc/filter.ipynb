```python
import lib.filters as filters
import lib.beat_util as util
import matplotlib.pyplot as plt
%matplotlib widget

sr = 22050
```

# Plot the filters (individually)

## Sub (frequency band) filter

```python
b, a = filters.create_sub_filter(sr)
plt.figure()
util.plot_filter_response(a, b, sr, use_log_scale=True, title="Sub Filter")
```

## Low (frequency band) filter

```python
b, a = filters.create_low_filter(sr)
plt.figure()
util.plot_filter_response(a, b, sr, use_log_scale=True, title="Low Filter")
```

## Midrange (frequency band) filter

```python
b, a = filters.create_midrange_filter(sr)
plt.figure()
util.plot_filter_response(a, b, sr, use_log_scale=True, title="Midrange Filter")
```

## High midrange (frequency band) filter

```python
b, a = filters.create_high_midrange_filter(sr)
plt.figure()
util.plot_filter_response(a, b, sr, use_log_scale=True, title="High Midrange Filter")
```

## High (frequency band) filter

```python
b, a = filters.create_high_filter(sr)
plt.figure()
util.plot_filter_response(a, b, sr, use_log_scale=True, title="High Filter")
```

# Plot all the filters (in the same graph)

```python
b_sub, a_sub = filters.create_sub_filter(sr)
b_low, a_low = filters.create_low_filter(sr)
b_mid, a_mid = filters.create_midrange_filter(sr)
b_hmid, a_hmid = filters.create_high_midrange_filter(sr)
b_high, a_high = filters.create_high_filter(sr)

b = [b_sub, b_low, b_mid, b_hmid, b_high]
a = [a_sub, a_low, a_mid, a_hmid, a_high]

plt.figure()
util.plot_all_filter_response(a, b, sr, use_log_scale=True, title="Filters")
plt.legend(["Sub", "Low", "Mid", "High Mid", "High"], loc="lower right");
```
